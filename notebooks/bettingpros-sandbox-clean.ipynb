```python
features = [
    'season',  # season
    ## Stats for home team going into game
    # Games 1 -> N-1
    'home_games_played', # total games played in season
    'home_win_pct', # winning percentage
    
    #'home_runs_for_mean', # mean runs scored
    #'home_runs_against_mean', # mean runs allowed
    
    'home_run_differential_mean', # mean run differential
    
    #'home_runs_for_std', # standard deviation of runs scored 
    #'home_runs_against_std', # standard deviation of runs allowed
    
    'home_run_differential_std', # standard deviation of run differential
    'home_win_streak', # streak of wins going into game (0 if most recently lost)
    'home_loss_streak', #streak of losses going into game (0 if most recently won)
    ## summary stats for last 5 and 10 games
    
    #'home_runs_for_mean_last_5', 
    #'home_runs_for_mean_last_10',
    #'home_runs_for_std_last_5', 
    #'home_runs_for_std_last_10',
    #'home_runs_against_mean_last_5', 
    #'home_runs_against_mean_last_10',
    #'home_runs_against_std_last_5', 
    #'home_runs_against_std_last_10',
    
    'home_run_differential_mean_last_5',
    'home_run_differential_mean_last_10',
    'home_run_differential_std_last_5', 
    'home_run_differential_std_last_10',
    'home_win_pct_last_5', 
    'home_win_pct_last_10', 
    ## Equivalent stats for away team
    'away_games_played',
    'away_win_pct', 
    
    #'away_runs_for_mean', 
    #'away_runs_against_mean',
    
    'away_run_differential_mean', 
    
    #'away_runs_for_std',
    #'away_runs_against_std', 
    
    'away_run_differential_std', 
    'away_win_streak',
    'away_loss_streak', 
    
    #'away_runs_for_mean_last_5',
    #'away_runs_for_mean_last_10', 
    #'away_runs_for_std_last_5',
    #'away_runs_for_std_last_10', 
    #'away_runs_against_mean_last_5',
    #'away_runs_against_mean_last_10', 
    #'away_runs_against_std_last_5',
    #'away_runs_against_std_last_10', 
    
    'away_run_differential_mean_last_5',
    'away_run_differential_mean_last_10',
    'away_run_differential_std_last_5', 
    'away_run_differential_std_last_10',
    'away_win_pct_last_5', 
    'away_win_pct_last_10', 
    'minutes_pacific', # gametime represented in minutes since start of day in PST

    'humidity', 
    'pressure', 
    'precipitation_chance', 
    'wind_speed',
    'wind_degree',
    'temperature',
    'home_moneyline',
    'away_moneyline',
    'home_spread',
    'away_spread',
    'home_spread_odds',
    'away_spread_odds',
    'over_under',
    'over_odds',
    'under_odds',
]

total_score = (samples_df['home_score']+samples_df['away_score']) - samples_df['over_under']
evaluation_metric = 'point_diff_less_spread'
samples_df['point_diff_less_spread'] = samples_df['home_score'] - samples_df['away_score'] + samples_df['home_spread']
# Plot a bar plot where the height of each bar is the correlation coefficient betwee each feature in features and total_score; excluding nans from the calculation
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
corr = samples_df[features+[evaluation_metric]].corr()
corr = corr.loc[features, evaluation_metric]
corr = corr[~np.isnan(corr)]
corr = corr.sort_values()
plt.figure(figsize=(10,10))
sns.barplot(x=corr.values, y=corr.index)
plt.title(f'Correlation between features and {evaluation_metric}')
plt.xlabel('Correlation')
plt.ylabel('Feature')
plt.show()

```

```python
from sklearn.model_selection import train_test_split
sample_df = samples_df.dropna(subset=features+[evaluation_metric]).copy()
train_df, test_df = train_test_split(sample_df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_df, test_size=0.2, random_state=42)
print(f'Training set shape: {train_df.shape}')

# Rescale features to have mean 0 and standard deviation 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_df[features] = scaler.fit_transform(train_df[features])
val_df[features] = scaler.transform(val_df[features])
test_df[features] = scaler.transform(test_df[features])

# Perform Linear regression with sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(train_df[features], train_df[evaluation_metric])
train_preds = lr.predict(train_df[features])
val_preds = lr.predict(val_df[features])
test_preds = lr.predict(test_df[features])
print(f'Training MSE: {mean_squared_error(train_df[evaluation_metric], train_preds)}')
print(f'Validation MSE: {mean_squared_error(val_df[evaluation_metric], val_preds)}')
print(f'Test MSE: {mean_squared_error(test_df[evaluation_metric], test_preds)}')
```

```python
def plot_regression_scatter(temp_df, model, sample_type: str):
    observed = temp_df[evaluation_metric].values
    #prediction = lr.predict(temp_df[features])
    prediction = model.predict(temp_df[features])
    #observed = np.sign(observed)*np.exp(np.abs(observed))
    #prediction = np.sign(prediction)*np.exp(np.abs(prediction))

    # plot the prediction vs action for the training set and validation set add histograms of each in side and top panels
    fig, ax = plt.subplots(2, 2, figsize=(8, 8), width_ratios=[4,1], height_ratios=[1, 4])
    # Plot histogram of evaluation metric in top left
    ax[0][0].set_title(f'Score Differential ({sample_type})')
    ax[0][0].hist(observed, color='midnightblue', bins=np.arange(-15.5, 15.5, 1))
    ax[0][0].set_xlim(-15, 15)

    # Plot scatter plot of evaluation metric vs prediction in bottom left
    ax[1][0].plot(observed, prediction, 'o', color='midnightblue', alpha=0.1)
    ax[1][0].axhline(0, color='black', linestyle='--')
    ax[1][0].axvline(0, color='black', linestyle='--')
    ax[1][0].set_ylabel('Frequency', fontsize=15)
    ax[1][0].set_xlabel('Observed', fontsize=15)
    ax[1][0].set_ylabel('Predicted')
    # add line of best fit
    ps = np.polyfit(observed, prediction, 1)
    x = np.linspace(-15, 15, 100)
    y = ps[0]*x + ps[1]
    ax[1][0].plot(x, y, 'c')
    ax[1][0].set_xlim(-15, 15)
    ax[1][0].set_ylim(-15, 15)
    # Add correlation coefficient to text on plot
    corr = np.corrcoef(observed, prediction)[0][1]
    ax[1][0].text(-14, 13, f'Correlation: {corr:.2f}', fontsize=15)
    # plot histogram of predictions transposed in bottom right
    ax[1][1].hist(prediction, orientation='horizontal', color='midnightblue', bins=np.arange(-15.5, 15.5, 1))
    ax[0][1].spines.right.set_visible(False)
    ax[0][1].spines.left.set_visible(False)
    ax[0][1].spines.top.set_visible(False)
    ax[0][1].spines.bottom.set_visible(False)
    ax[0][1].tick_params(axis='both', which='both', bottom=False, top=False, labelbottom=False, right=False, left=False, labelleft=False)
    plt.show()

plot_regression_scatter(train_df, lr, 'Training')
plot_regression_scatter(val_df, lr, 'Validation')
plot_regression_scatter(test_df, lr, 'Test')
```

```python
market_id = 122 # 122 is moneyline
market_id = 175 # 175 is total runs
market_id = 176 # 176 is spread

selections = next(offer['selections'] for offer in line_data['offers'] if offer['market_id'] == market_id)

from typing import Dict, Any, List
class GameBets:

    def __init__(self, line_data: Dict[str, Any]):
        self.line_data = line_data
        self.market_dict = {
            122: 'moneyline',
            175: 'total runs',
            176: 'spread'
        }
        self.market_ids = list(self.market_dict.keys())
        

    def parse_selection(self, selection: List[Dict[str, Any]]):
        bet = selection['selection'] if market_id == 175 else selection['participant']
        dk_selection = next(line for line in selection['books'] if line['id'] == 12)
        odds = dk_selection['lines'][0]['cost']
        line = dk_selection['lines'][0]['line']
        return {
            'market_id': market_id,
            'bet_type': self.market_dict[market_id],
            'bet': bet,
            'odds': odds,
            'line': line
        }

    def parse_market(self, market_id: int):
        selections = next(offer for offer in self.line_data['offers'] if offer['market_id'] == market_id)['selections']
        return [self.parse_selection(selection) for selection in selections]

    def parse_markets(self):
        return {self.market_dict[market_id]: self.parse_market(market_id) for market_id in self.market_ids}
output = GameBets(line_data).parse_markets()
output
```
